Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
#happiness_df['happiness_degree'] =np.nan
pd.set_option('mode.chained_assignment',None)
happiness_df = happiness_df.assign(happiness_degree='unhappy')
for i in happiness_df['Ladder']:
    if happiness_df['Ladder'][i-1]<30:
        happiness_df['happiness_degree'][i-1] = "very happy"
    elif 30<=happiness_df['Ladder'][i-1]<60:
        happiness_df['happiness_degree'][i-1] = "happy"
    elif 60<=happiness_df['Ladder'][i-1]<90:
        happiness_df['happiness_degree'][i-1] = "moderately happy"
    elif 90<=happiness_df['Ladder'][i-1]<120:
        happiness_df['happiness_degree'][i-1] = "unhappy"
    else:
        happiness_df['happiness_degree'][i-1] = "very unhappy"  
happiness_df.head(10)
```

Question 10) Draw random samples 30%, 40%, 50%, 60% of the data with replacement and investigate how the correlation of sample to the whole data changes with respect to sample size. (15 points)

```python
#codes
# I just wanted to add 70-90-95 cause I wanted to see how correlation changes and how pie charts changes.
dfs30 = happiness_df.sample(frac = 0.3, replace = True, random_state = 1)
display(dfs30.corr())
```

```python
dfs40 = happiness_df.sample(frac = 0.4, replace = True, random_state = 1)
display(dfs40.corr())
```

```python
dfs50 = happiness_df.sample(frac = 0.5, replace = True, random_state = 1)
display(dfs50.corr())
```

```python
dfs60 = happiness_df.sample(frac = 0.6, replace = True, random_state = 1)
display(dfs60.corr())
```

Question 11) Plot pie charts for random sampled data obtained in previous question according to happiness_degree. (10 points)

```python
#codes

group_dfs30 = dfs30.groupby('happiness_degree')['happiness_degree'].agg('count')

group_dfs30.plot.pie(subplots = True, autopct='%.2f')
plt.title('Happiness Degree', fontsize=20, loc='center')
plt.ylabel('')
```
